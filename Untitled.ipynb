```python
import pandas as pd

# Load dataset
df = pd.read_csv("sales_data 3.csv")

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Drop rows where total sales are missing or zero
df.dropna(subset=['total'], inplace=True)
df = df[df['total'] > 0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

```

```python
top_products = df.groupby('item_name')['total'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index, palette='Blues_r')
plt.title("Top 10 Selling Products")
plt.xlabel("Total Sales (Tk)")
plt.ylabel("Product")
plt.tight_layout()
plt.savefig("top_products.png")
plt.show()
```

```python
category_sales = df.groupby('category')['total'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=category_sales.values, y=category_sales.index, palette='coolwarm')
plt.title("Sales by Category")
plt.xlabel("Total Sales (Tk)")
plt.ylabel("Category")
plt.tight_layout()
plt.savefig("sales_by_category.png")
plt.show()
```

```python
brand_sales = df.groupby('brand')['total'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=brand_sales.values, y=brand_sales.index, palette='mako')
plt.title(" Top Brands by Sales")
plt.xlabel("Total Sales (Tk)")
plt.ylabel("Brand")
plt.tight_layout()
plt.savefig("sales_by_brand.png")
plt.show()
```

```python
employee_sales = df.groupby('employee_name')['total'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=employee_sales.values, y=employee_sales.index, palette='viridis')
plt.title("Top Employees by Sales")
plt.xlabel("Total Sales (Tk)")
plt.ylabel("Employee")
plt.tight_layout()
plt.savefig("sales_by_employee.png")
plt.show()
```
